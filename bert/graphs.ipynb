```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import Pandas and Matplotlib
# If not installed, you can install using: pip install pandas matplotlib

# Step 2: Load CSV into DataFrame
file_path = 'training data.csv'
df = pd.read_csv(file_path, sep=';')

# Step 3: Plot specific columns
plt.figure(figsize=(10, 10))  # Adjust the figure size if necessary

# Plot Temperature
```

```python
df
```

```python
plt.plot(df['Step'], df['Accuracy'], marker='o', linestyle='-', color='g', label='Accuracy')

# Plot Humidity
plt.plot(df['Step'], df['F1'], marker='s', linestyle='--', color='r', label='F1')
plt.plot(df['Step'], df['Precision'], marker='o', linestyle='-', color='b', label='Precision')

# Plot Humidity
plt.plot(df['Step'], df['Recall'], marker='s', linestyle='--', color='black', label='Recall')

# Customize plot labels and title
plt.xlabel('Step')
# plt.ylabel('Step')
plt.title('Accuracy, F1, Precision and Recall over time')
plt.legend()

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.plot(df['Step'], df['Training Loss'], marker='o', linestyle='-', color='b', label='Training Loss')

# Plot Humidity
plt.plot(df['Step'], df['Validation Loss'], marker='s', linestyle='--', color='r', label='Validation Loss')

# Customize plot labels and title
plt.xlabel('Step')
# plt.ylabel('Step')
plt.title('Training Loss and Validation Loss over time')
plt.legend()

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
